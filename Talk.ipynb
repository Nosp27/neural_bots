```python
import src.net as net
from src.net import MessageBot
import pandas as pd
from time import sleep
import random
```

```python
straight_history_df = pd.read_table('models/k_model/lstm_in.txt', names=['me', 'other', 'rest'])
inversed_history_df = pd.read_table('./models/k_model/lstm_in.txt', names=['other', 'me', 'rest'])
```

```python
my_model = MessageBot('models/k_model/my_k_model.h5', history_df=inversed_history_df)
other_model = MessageBot(
    'models/k_model/model_weights.h5', 
    history_df=straight_history_df
)
```

```python
my_model.translate('да')
```

```python
def impr_tr(messenger, wrd):
    try:
        return messenger.translate(wrd)
    except KeyError:
        return messenger.translate(wrd.replace('ё', 'е'))

stop_combination = 'в чс'
wp = ''
word2 = 'Писька'
for i in range(20):
    print('X: %s' % word2)
    word1 = impr_tr(my_model, word2)
    print('Me: %s' % word1)
    sleep(2.4)
    if word2 == stop_combination:
        word2 = straight_history_df.other[random.randint(0, straight_history_df.shape[0] - 1)]
    else:
        word2 = impr_tr(other_model, word1)
    if word2 == wp:
        print('X: %s' % stop_combination)
        print()
        print()
        his = other_model.lines
        word2 = his.other[random.randint(0, his.shape[0] - 1)]
    wp = word2
    sleep(2.4)
```
